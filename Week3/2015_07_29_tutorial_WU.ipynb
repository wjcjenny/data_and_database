##### A real-world example: Columbia Journalism Faculty

<img src="li_tag.png">

```python
url = 'http://www.journalism.columbia.edu/page/10/10?category_ids%5B%5D=2&category_ids%5B%5D=3&category_ids%5B%5D=37'
from bs4 import BeautifulSoup
import urllib
```

```python
html_str = urllib.urlopen(url).read()
```

```python
print html_str[:500]
```

```python
document = BeautifulSoup(html_str)
```

```python
li_tags = document.find_all('li')
for tag in li_tags:
    print "here is a tag:", tag
```

```python
ul_tag = document.find('ul', attrs = {'class': 'experts-list'})
print ul_tag
```

```python
li_tags = ul_tag.find_all('li')
```

```python
all_faculty = []
for tag in li_tags:
    #print "here is a tag:", tag
    faculty_member = {}
    h4_tag = tag.find('h4')
    if h4_tag is not None:
        a_tag = h4_tag.find('a')
        faculty_name = a_tag.string
        p_tag = tag.find('p',attrs = {'class': 'description'})
        if p_tag is not None:
            faculty_title = p_tag.string
            faculty_member = {'name': faculty_name,'title':faculty_title}
            all_faculty.append(faculty_member)
all_faculty
```

```python
# import sys  
# reload(sys)  
# sys.setdefaultencoding('utf8')

import pandas as pd
faculty_frame = pd.DataFrame(all_faculty)
faculty_frame[faculty_frame['title'] =="Adjunct Faculty"]
```
